```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer, MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("anime.csv") 

# Inspect data
df.head()
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df['genre'] = df['genre'].fillna('Unknown')
df['genre'] = df['genre'].apply(lambda x: x.split(', '))
```

```python
mlb = MultiLabelBinarizer()
genre_encoded = pd.DataFrame(mlb.fit_transform(df['genre']), columns=mlb.classes_)
genre_encoded
```

```python
# Normalize rating and members
scaler = MinMaxScaler()
numerical_features = df[['rating', 'members']].fillna(0)
numerical_scaled = pd.DataFrame(scaler.fit_transform(numerical_features), columns=['rating', 'members'])
numerical_scaled
```

```python
features = pd.concat([genre_encoded, numerical_scaled], axis=1)
features
```

```python
features.drop(columns='Unknown')
```

```python
similarity_matrix = cosine_similarity(features)
```

```python
def recommend_anime(title, top_n=5, threshold=0.5):
    idx = df[df['name'] == title].index[0]
    sim_scores = list(enumerate(similarity_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = [i for i in sim_scores if i[1] >= threshold and i[0] != idx][:top_n]
    anime_indices = [i[0] for i in sim_scores]
    return df['name'].iloc[anime_indices]
```
